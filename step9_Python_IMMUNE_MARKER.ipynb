```python
def file_parsing_mfa(data):

    data = data.reset_index(drop=False)
    data = data.rename(columns={'index':'gene_chr'})
    data_melt = pd.melt(data, id_vars=['gene_chr'])
    conditionlist = [data_melt.variable == 'A_FST',
                     data_melt.variable == 'B_FST',
                     data_melt.variable == 'A_SND',
                     data_melt.variable == 'B_SND',
                     data_melt.variable == 'A_TRD',
                     data_melt.variable == 'B_TRD',
                 
                     data_melt.variable == 'C_FST',
                     data_melt.variable == 'D_FST',
                     data_melt.variable == 'C_SND',
                     data_melt.variable == 'D_SND',
                     data_melt.variable == 'C_TRD',
                     data_melt.variable == 'D_TRD',
                 
                     data_melt.variable == 'E_FST',
                     data_melt.variable == 'F_FST',
                     data_melt.variable == 'E_SND',
                     data_melt.variable == 'F_SND',
                     data_melt.variable == 'E_TRD',
                     data_melt.variable == 'F_TRD',
                 
                     data_melt.variable == 'G_FST',
                     data_melt.variable == 'H_FST',
                     data_melt.variable == 'G_SND',
                     data_melt.variable == 'H_SND',
                     data_melt.variable == 'G_TRD',
                     data_melt.variable == 'H_TRD']
    choicelist = [3,3,4,4,5,5,7,7,8,8,9,9,12,12,13,13,14,14,16,16,17,17,18,18]
    data_melt['age'] = np.select(conditionlist,choicelist, default='Not Specified')


    conditionlist = [data_melt.variable.str.split('_').str[0] == 'A',
                     data_melt.variable.str.split('_').str[0] == 'B',
                     data_melt.variable.str.split('_').str[0] == 'C',
                     data_melt.variable.str.split('_').str[0] == 'D',
                     data_melt.variable.str.split('_').str[0] == 'E',
                     data_melt.variable.str.split('_').str[0] == 'F',
                     data_melt.variable.str.split('_').str[0] == 'G',
                     data_melt.variable.str.split('_').str[0] == 'H']
    choicelist = ['G1','G1','G2','G2','G3','G3','G4','G4']
    data_melt['Group'] = np.select(conditionlist,choicelist, default='Not Specified')

    data_melt['value'] = data_melt['value'].round()
    data_melt['value'] = pd.to_numeric(data_melt['value'])
    data_melt['age'] = pd.to_numeric(data_melt['age'])
    
    return data_melt


def file_parsing_mml(data):
    
    GeneCPM_MML = data.reset_index(drop=False)
    GCL = pd.melt(GeneCPM_MML, id_vars=['gene_chr'])
    GCL = GCL.rename(columns={'variable':'sample'})
    GCL = pd.merge(GCL,condition,how='inner',on='sample')
    
    conditionlist = [(5 >= GCL['age']) & (GCL['age'] >= 4),
                 (11 >= GCL['age']) & (GCL['age'] >= 7),
                 (19 >= GCL['age']) & (GCL['age'] >= 15),
                 (22 >= GCL['age']) & (GCL['age'] >= 21)]
    choicelist = ['G1','G2','G3','G4']
    GCL['Group'] = np.select(conditionlist,choicelist, default='Not Specified')
                             
    return GCL

def zs(df):

    DF = pd.DataFrame()
    for i in df['gene'].unique():
        G = df[df['gene'] == i]       
        G = G.copy()
        G['z'] = stats.zscore(G['value'])
        DF = pd.concat([DF,G], axis=0)
        
    return DF
```

```python
mfa_ = file_parsing_mfa(DEG_norm_count_mfa)
mfa_['gene'] = mfa_.gene_chr.str.split('_').str[0]
ZS_CPM = zs(mfa_)
```

```python
import statannot
import scikit_posthocs as sp
import matplotlib.gridspec as gridspec

def each_Graph(df,cell_type,gspec,N,Type):

    dt = df[df['Cell Type'] == cell_type]
    dt = dt.copy()
    dt['group_w'] = dt['group_w'].astype('str')
    dt = dt.reset_index(drop=True)

    Col = ['lightgrey' if dt.loc[idx]['group_w'] == 'Not' 
           else C1 if dt.loc[idx]['group_w'] == 'PT_1'
           else C2 if dt.loc[idx]['group_w'] == 'PT_2'
           else C3 if dt.loc[idx]['group_w'] == 'PT_3'
           else C4 for idx in dt.index]

    plt.subplot(gspec[0,N])
    plt.xticks([])
    plt.yticks([])
    plt.text(0.5, 0.5, cell_type, ha='center', va='center', size=10)
    plt.gca().set_facecolor('peachpuff') #배경색
        
    globals()['ax{}'.format(N)] = plt.subplot(gspec[1,N])
        
    if Type == 'dot':
        ax1 = sns.stripplot(data = dt, x='Group', y='z',hue=dt.index,palette=Col,jitter=True, alpha=.7)
    if Type == 'violin':
        ax1 = sns.violinplot(data = dt, x='Group', y='value',alpha=.7,palette=['#ED8080','#90EE90','#ADD8E6','#EFBBF0'],showfliers=False) 
        if cell_type == 'B cell' or cell_type == 'Helper T cell':
            statannot.add_stat_annotation(ax1, data = dt, x='Group', y='value', order = ['G1','G2','G3','G4'],
                                          box_pairs = [('G3','G4'),('G2','G3'),('G1','G2'),('G1','G4')],
                                          test='Mann-Whitney',text_format='star',loc='inside',verbose=2)
        else:
            statannot.add_stat_annotation(ax1, data = dt, x='Group', y='value', order = ['G1','G2','G3','G4'],
                                          box_pairs = [('G1','G2'),('G2','G3'),('G3','G4'),('G1','G4')],
                                          test='Mann-Whitney',text_format='star',loc='inside',verbose=2)
    plt.yticks(fontsize=8)
    plt.xticks(fontsize=13)
    ax1.set(xlabel=None)
    ax1.set(ylabel=None)
        
    if Type == 'dot':
        ax1.get_legend().remove()
        


def Graph(T_df,T_name,C_df,C_name,Type):

    fig = plt.figure(figsize = (20,10))

    gs = gridspec.GridSpec(4,1) 
    
    gs0 = gridspec.GridSpecFromSubplotSpec(nrows=2, ncols=8, height_ratios=[0.5,2],width_ratios=[0.2,1,1,1,1,1,1,1],hspace=0,wspace=0.4, subplot_spec=gs[0])
    gs1 = gridspec.GridSpecFromSubplotSpec(nrows=2, ncols=8, height_ratios=[0.5,2],width_ratios=[0.2,1,1,1,1,1,1,1],hspace=0,wspace=0.4, subplot_spec=gs[1])


    plt.subplot(gs0[0:2,0])
    plt.xticks([])
    plt.yticks([])
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    if Type == 'dot':
        plt.text(1, 0.5, 'Macaca fascicularis z-score\nof nomalized counts', ha='center', va='center', size=10,  rotation=90)
    if Type == 'violin':
        plt.text(0.8, 0.5, 'Macaca fascicularis\nof nomalized counts', ha='center', va='center', size=12,  rotation=90)
    
    plt.subplot(gs1[0:2,0])
    plt.xticks([])
    plt.yticks([])
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    if Type == 'dot':
        plt.text(1, 0.5, 'Macaca mulatta z-score\nof nomalized counts', ha='center', va='center', size=10,  rotation=90)
    if Type == 'violin':
        plt.text(0.8, 0.5, 'Macaca fascicularis\nof nomalized counts', ha='center', va='center', size=12,  rotation=90)
    
    T_df = T_df.sort_values(by=['Cell Type','Group'], ascending=[True,True])
    C_df = C_df.sort_values(by=['Cell Type','Group'], ascending=[True,True])

    a = 1
    for im in T_df['Cell Type'].unique():
        each_Graph(T_df,im,gs0,a,Type)
        a += 1
    b = 1
    for im in C_df['Cell Type'].unique():
        each_Graph(C_df,im,gs1,b,Type)
        b += 1
  
    plt.rcParams["figure.autolayout"] = True
    plt.tight_layout()
#    if Type == 'dot':
#        plt.savefig("/home/chm/analyzing_file_byR/RNA/SQuIRE_count_ELTC/figures/immune_marker_mfa_mml_dot.png",dpi=1000)
#    else:
#        plt.savefig("/home/chm/analyzing_file_byR/RNA/SQuIRE_count_ELTC/figures/immune_marker_mfa_mml_violin.png",dpi=1000)
    plt.show()

Graph(mfa_imm,'mfa',mml_imm,'mml','dot')
Graph(mfa_imm,'mfa',mml_imm,'mml','violin')
```

```python
from upsetplot import UpSet, from_indicators, generate_counts


def upsetplot(df):
    
    df = df.dropna(axis=0)
    DA = df[['G','gene']]
    DA = DA.dropna(axis=0)
    DA = DA.set_index('gene')

    all_G = DA['G'].unique().tolist()
    all_G.reverse()
    all_gene = DA.index.unique().tolist()
    
    A = DA[DA['G']==all_G[0]]
    B = DA[DA['G']==all_G[1]]
    C = DA[DA['G']==all_G[2]]
    D = DA[DA['G']==all_G[3]]
    E = DA[DA['G']==all_G[4]]
    F = DA[DA['G']==all_G[5]]
    G = DA[DA['G']==all_G[6]]
    H = DA[DA['G']==all_G[7]]
    I = DA[DA['G']==all_G[8]]
    J = DA[DA['G']==all_G[9]]
    K = DA[DA['G']==all_G[10]]
    L = DA[DA['G']==all_G[11]]
    M = DA[DA['G']==all_G[12]]
    N = DA[DA['G']==all_G[13]]
    O = DA[DA['G']==all_G[14]]
    P = DA[DA['G']==all_G[15]]
    Q = DA[DA['G']==all_G[16]]
    R = DA[DA['G']==all_G[17]]

    TOT = pd.concat([A,B,C,D,E,F,G,H,I,J,K,L,M,N,O,P,Q,R],axis=1)
    TOT.columns=all_G

    TOT = TOT.notnull()

    TOT_G = TOT.groupby(all_G).size()

    #DEG
    DEG_upset = UpSet(TOT_G,show_counts=True, show_percentages=False,orientation='horizontal',sort_categories_by=None,facecolor='black')
    DEG_upset.style_subsets(present=['PT_2','TC_Pattern3','B cell'],facecolor=C2)
    DEG_upset.style_subsets(present=['PT_2','TC_Pattern3','Helper T cell'],facecolor=C2)
    DEG_upset.style_subsets(present=['PT_4','TC_Pattern6','B cell'],facecolor=C4)
    DEG_upset.style_subsets(present=['PT_1','TC_Pattern2','Cytotoxic T cell'],facecolor=C1)
    DEG_upset.style_subsets(present=['PT_1','TC_Pattern2','NK cell'],facecolor=C1)
    DEG_upset.style_subsets(present=['PT_1','TC_Pattern2','CD14+ Monocyte'],facecolor=C1)
    DEG_upset.style_subsets(present=['PT_1','TC_Pattern2','CD14- Monocyte'],facecolor=C1)
    DEG_upset.style_subsets(present=['PT_3','TC_Pattern4','CD14+ Monocyte'],facecolor=C3)
    DEG_upset.style_subsets(present=['PT_3','TC_Pattern4','CD14- Monocyte'],facecolor=C3)

    DEG_upset.plot()
    plt.suptitle('Macaca fascicularis',x=0.55,y=1,fontsize=20)
#    plt.savefig("/home/chm/analyzing_file_byR/RNA/SQuIRE_count_ELTC/figures/DEG_WGCNA_cell_type_intersection.png",dpi=900)
    plt.show()
    plt.show()

immune_mark_m = immune_mark_cyno.rename(columns={'Cell Type':'G','Gene':'gene'})
    
DEG_ALL_m = DEG_ALL[['gene','group']]
DEG_ALL_m = DEG_ALL_m.rename(columns={'group':'G'})

Pattern3_m = Patterns[['gene','group_w']]
Pattern3_m = Pattern3_m.rename(columns={'group_w':'G'})
sel_DF = pd.concat([immune_mark_m,DEG_ALL_m,Pattern3_m],axis=0)

upsetplot(sel_DF)
```
