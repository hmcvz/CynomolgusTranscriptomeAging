# HEATMAP

```python
def heatmap(df,df_GO,df_GOAll,CMAP,N):
    df_Name = df.name
    df_GO = df_GO[['GO','Log(q-value)']]
    DF = pd.merge(df,df_GO,how='inner', on='GO')
    DF_each = df_GOAll[df_GOAll['GO'].isin(df['GO'])]
    
    en = DF_each[['GO','Enrichment','GeneList']]
    nu = DF_each[['GO','#GeneInGOAndHitList','GeneList']]
    en = en.pivot(index=['GO'],columns=['GeneList'],values='Enrichment')
    nu = nu.pivot(index=['GO'],columns=['GeneList'],values='#GeneInGOAndHitList')
    
    en = en.reset_index(drop=False)
    nu = nu.reset_index(drop=False)
    
    en = en.rename(columns={'human_group1':'enr1','human_group2':'enr2','human_group3':'enr3',
                            'human_group4':'enr4','human_group6':'enr6','human_group7':'enr7',
                            'greenyellow':'enr1','turquoise':'enr2',
                            'human_WGCNA_group2':'enr2','human_WGCNA_group3':'enr3','human_WGCNA_group6':'enr6',
                            'PT_1':'enr2','PT_2':'enr3','PT_4':'enr6',
                            'human_G1_G2_up':'enr1','human_G1_G2_down':'enr2','human_G2_G3_up':'enr3',
                            'human_G2_G3_down':'enr4','human_G3_G4_up':'enr5','human_G3_G4_down':'enr6'})
    nu = nu.rename(columns={'human_group1':'nu1','human_group2':'nu2','human_group3':'nu3',
                            'human_group4':'nu4','human_group6':'nu6','human_group7':'nu7',
                            'greenyellow':'nu1','turquoise':'nu2',
                            'human_WGCNA_group2':'nu2','human_WGCNA_group3':'nu3','human_WGCNA_group6':'nu6',
                            'PT_1':'nu2','PT_2':'nu3','PT_4':'nu6',
                            'human_G1_G2_up':'nu1','human_G1_G2_down':'nu2','human_G2_G3_up':'nu3',
                            'human_G2_G3_down':'nu4','human_G3_G4_up':'nu5','human_G3_G4_down':'nu6'})
    
    NUM = len(en.columns)-1
    EN_NU = pd.merge(en,nu,how='inner', on='GO')
    

    DF = pd.merge(DF,EN_NU,how='inner', on='GO')
    DF = DF.sort_values(by=['Log(q-value)'], ascending=[True])

    for i in DF.columns:
        if i.startswith('_LogP'):
            DF[i] = abs(DF[i])
            if DF[i].sum() == 0:
                DF = DF.loc[:,~DF.columns.str.contains(i)]
            
    DF = DF.replace(np.nan,0)
    DF = DF.reset_index(drop=False)
    DF = DF.sort_values(by=['index'], ascending=[True])

    
    fig = plt.figure(figsize = (NUM,10))

    gs = gridspec.GridSpec(nrows=3, ncols=1, height_ratios=[1,0.4,40],hspace=0.05)

    ax1 = plt.subplot(gs[0,0])
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)

    ax2 = plt.subplot(gs[2,0])
    plt.yticks(fontsize=20)
    

    
    #ax1
    cate = range(NUM)
    group = range(NUM)
    df = pd.DataFrame({'group': group, 'cate': cate })
    df_heatmap = df.pivot_table(values='cate',columns='group')
    sns.heatmap(df_heatmap,cmap=CMAP,annot=False, ax=ax1, cbar=False)
    ax1.axes.get_xaxis().set_visible(False)
    ax1.axes.get_yaxis().set_visible(False)

    #ax2
    DF_h = DF.set_index('Description')
    DF_h = DF_h.loc[:,DF_h.columns.str.contains('enr')]
    cbar_ax = fig.add_axes([.92,.3,.02,.4])
    #sns.heatmap(DF_h,linewidths=2,annot=False, ax=ax2, cbar=True,cmap='Reds', linecolor='grey')
    sns.heatmap(DF_h,linewidths=2,annot=False, ax=ax2, cmap='Reds', linecolor='grey', cbar_ax=cbar_ax)
    ax2.axes.get_xaxis().set_visible(False)
    
    #ax3
    if df_Name == 'PW':
        
        
        #ax1
        cate = range(NUM)
        group = range(NUM)
        df = pd.DataFrame({'group': group, 'cate': cate })

        df_heatmap = df.pivot_table(values='cate',columns='group')

        sns.heatmap(df_heatmap,cmap=CMAP,annot=False, ax=ax1, cbar=False)
        ax1.axes.get_xaxis().set_visible(False)
        ax1.axes.get_yaxis().set_visible(False)
        
        #ax3
        ax3 = plt.subplot(gs[1,0])
        plt.gca().spines['right'].set_visible(False)
        plt.gca().spines['left'].set_visible(False)
        plt.gca().spines['top'].set_visible(False)
        plt.gca().spines['bottom'].set_visible(False)
        
        cate = range(NUM*2)
        group = range(NUM*2)
        df = pd.DataFrame({'group': group, 'cate': cate })

        df_heatmap = df.pivot_table(values='cate',columns='group')

        sns.heatmap(df_heatmap,cmap=['red','blue','red','blue','red','blue'],annot=False, ax=ax3, cbar=False)
        ax3.axes.get_xaxis().set_visible(False)
        ax3.axes.get_yaxis().set_visible(False)

    
    return DF
 
    
    
TC_cmap=['darksalmon','sandybrown','goldenrod','lightgreen','lightblue','violet']
WGCNA_cmap=['greenyellow','turquoise']
WGCNA_TC_cmap=['#66c2a5','#8da0cb','#deb887']
PW_cmap=['#BFB788','#9FE3BB','#CECAEB']
    
TC_DF = heatmap(TC,TC_GO,TC_GOAll,TC_cmap,'TC')
WGCNA_DF =heatmap(WGCNA,WGCNA_GO,WGCNA_GOAll,WGCNA_cmap,'WGCNA')
WGCNA_TC_DF =heatmap(WGCNA_TC,WGCNA_TC_GO,WGCNA_TC_GOAll,WGCNA_TC_cmap,'WGCNA_TC')
HUB_DF =heatmap(HUB,HUB_GO,HUB_GOAll,WGCNA_TC_cmap,'HUB')
PW_DF =heatmap(PW,PW_GO,PW_GOAll,PW_cmap,'PW')
```

# BARPLOT

```python
def parseforbar(DF):
    DF['new_y'] = 19 - DF['index']
    DF = DF.set_index(['new_y', 'GO', 'Description'])

    LogP = DF.loc[:,DF.columns.str.contains('_LogP')]
    nu = DF.loc[:,DF.columns.str.contains('nu')]
    
    LogP = LogP.reset_index(drop=False)
    nu = nu.reset_index(drop=False)
    
    LogP_melt = pd.melt(LogP, id_vars=['new_y', 'GO', 'Description'], var_name='LogP',value_name='LogP_value')
    nu_melt = pd.melt(nu , id_vars=['new_y', 'GO', 'Description'], var_name='nu',value_name='nu_value')
    
    df = pd.concat([LogP_melt,nu_melt[['nu','nu_value']]],axis=1)
    df['new_x'] = df.apply(lambda x: 1 if x['nu']=='nu1' else 2 if x['nu']=='nu2' else 3 , axis=1)

    return df


def bargraph(df,axes,clr,tex):
    
    sns.set_style('white')
    sns.barplot(x=df['LogP_value'], y=df['Description'],color=clr,ax=axes)
    sns.despine(top=True,left=False,bottom=False,right=True,ax=axes)

    rects = axes.patches
    labels = ["%s" % i for i in df['Description']]
    
    for rect, label in zip(rects, labels):
        
        width = rect.get_x() + 0.2
        height = rect.get_y() + 0.4
        axes.text(width,height,label,ha='left', va='center',size=12,color=tex)


    axes.legend().remove()
    axes.axes.get_yaxis().set_visible(False)
    
    if axes == ax3:
        axes.xaxis.set_label_text('-log10 pvalue',fontsize=15)
    else:
        axes.axes.get_xaxis().set_visible(False)
        #axes.xaxis.set_label_text('-log10 pvalue',fontsize=15)
```
