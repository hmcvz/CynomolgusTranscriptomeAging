```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
file = './source_data/cancer_incidence/cancer_diagnosis_bothsex.csv'
cancer_diag_both  = pd.read_csv(file, sep = ',', header = 4)

file = './source_data/cancer_incidence/cancer_diagnosis_male.csv'
cancer_diag_male = pd.read_csv(file, sep = ',', header = 4)

file = './source_data/cancer_incidence/cancer_diagnosis_female.csv'
cancer_diag_female  = pd.read_csv(file, sep = ',', header = 4)
```

```python
both = cancer_diag_both.iloc[1:20,:]
male = cancer_diag_male.iloc[1:20,:]
female = cancer_diag_female.iloc[1:20,:]

both = both.set_index('Unnamed: 0')
male = male.set_index('Unnamed: 0')
female = female.set_index('Unnamed: 0')

both = both.replace({'-' : '0.0'}, regex=True)
male = male.replace({'-' : '0.0'}, regex=True)
female = female.replace({'-' : '0.0'}, regex=True)

both = both.replace({',' : ''}, regex=True)
male = male.replace({',' : ''}, regex=True)
female = female.replace({',' : ''}, regex=True)


both = both.astype(float)
male = male.astype(float)
female = female.astype(float)
```

```python
BOTH = pd.DataFrame()
for i in range(0,177,3):
    y = both.iloc[:,i]
    BOTH = pd.concat([BOTH,y],axis=1)
    
MALE = pd.DataFrame()
for i in range(0,177,3):
    y = male.iloc[:,i]
    MALE = pd.concat([MALE,y],axis=1)
    
FEMALE = pd.DataFrame()
for i in range(0,177,3):
    y = female.iloc[:,i]
    FEMALE = pd.concat([FEMALE,y],axis=1)
```

```python
BOTH_P = pd.DataFrame()
for i in BOTH.columns:
    BOTH_P[i] = (BOTH[i] / BOTH[i].sum()) * 100
    
    
MALE_P = pd.DataFrame()
for i in MALE.columns:
    MALE_P[i] = (MALE[i] / MALE[i].sum()) * 100
    
    
FEMALE_P = pd.DataFrame()
for i in FEMALE.columns:
    FEMALE_P[i] = (FEMALE[i] / FEMALE[i].sum()) * 100
```

```python
BOTH_P = BOTH_P.reset_index()
BOTH_P = pd.melt(BOTH_P, id_vars=['index'])

MALE_P = MALE_P.reset_index()
MALE_P = pd.melt(MALE_P, id_vars=['index'])

FEMALE_P = FEMALE_P.reset_index()
FEMALE_P = pd.melt(FEMALE_P, id_vars=['index'])
```

```python
#for j in [BOTH_P,MALE_P,FEMALE_P]:
def MERGE_sel(j):    
    
    A = j[j['variable'].isin(['All Cancer Sites Combined'])]

#    Y = j[j['variable'].isin(['Glioblastoma of the Brain and ONS','Other Glioma of the Brain and ONS','Kidney','Liver',
#                              'Melanoma of the Skin'])]
    Y = j[j['variable'].isin(['Other Glioma of the Brain and ONS','Kidney','Liver','Colon and Rectum'])]

#    O = j[j['variable'].isin(['Squamous Cell Carcinoma of the Lung and Bronchus','Lung and Bronchus',
#                              'Cervix Uteri','Larynx'])]
    O = j[j['variable'].isin(['Squamous Cell Carcinoma of the Lung and Bronchus','Lung and Bronchus','Melanoma of the Skin',
                              'Cervix Uteri','Larynx'])]    
    
    #O = j[j['variable'].isin(['Squamous Cell Carcinoma of the Lung and Bronchus','Lung and Bronchus',
    #                          'HR+/HER2- Breast Cancer (Female only)','HR-/HER2+ Breast Cancer (Female only)',
    #                          'HR+/HER2+ Breast Cancer (Female only)','HR-/HER2- Breast Cancer (Female only)',
    #                          'Cervix Uteri'])]
        
    A['group'] = 'All combinded'
    Y['group'] = 'Young'
    O['group'] = 'Old'
    
    MERGE = pd.concat([A,Y,O],axis=0)
    
    A = A.rename(columns={'index':'Age','value':'Pertile','variable':'cancer types'})
    Y = Y.rename(columns={'index':'Age','value':'Pertile','variable':'cancer types'})
    O = O.rename(columns={'index':'Age','value':'Pertile','variable':'cancer types'})
    MERGE = MERGE.rename(columns={'index':'Age','value':'Pertile','variable':'cancer types'})
    
    return A,Y,O,MERGE
    
BOTH_A,BOTH_Y,BOTH_O,BOTH_M = MERGE_sel(BOTH_P)
MALE_A,MALE_Y,MALE_O,MALE_M = MERGE_sel(MALE_P)
FEMALE_A,FEMALE_Y,FEMALE_O,FEMALE_M = MERGE_sel(FEMALE_P)
```

```python
fig = plt.figure(figsize=(15,7))
gs = gridspec.GridSpec(nrows=2, ncols=2, height_ratios=[1,1],width_ratios=[1,19])

ax1 = fig.add_subplot(gs[0,0])
plt.xticks([])
plt.yticks([])
plt.text(2.2,0.23,'cancer_young',ha='right', size=15,rotation=90)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

ax3 = fig.add_subplot(gs[1,0])
plt.xticks([])
plt.yticks([])
plt.text(2.2,0.3,'cancer_old',ha='right', size=15,rotation=90)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)


ax2 = fig.add_subplot(gs[0,1])   
sns.lineplot(data=MALE_Y,x='Age',y='Pertile',hue='cancer types',palette='Set1',ax=ax2)
#ax2.get_legend().remove()
plt.xlabel('')
plt.ylabel('')
plt.ylim(0,30)
plt.xlim(0,18)

ax4 = fig.add_subplot(gs[1,1])  
sns.lineplot(data=MALE_O,x='Age',y='Pertile',hue='cancer types',palette='Set1',ax=ax4)
#ax4.get_legend().remove()
plt.ylabel('')
plt.ylim(0,30)
plt.xlim(0,18)

plt.show()
```
